## 1. read the data

```python
import pandas as pd

#load data
file_path = '20240415_noba_data_scientist_interview_task_dataset.csv'
data = pd.read_csv(file_path)

#summary
data.head(), data.describe(), data.info()
```

## 2. Preprocessing:

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

data.drop(columns=["application_id", "application_date"], inplace=True)

# Handle nan values
imputer = SimpleImputer(strategy="mean")
numerical_cols = [
    "year_gross_salary",
    "application_hour",
    "house_hold_mortgage",
    "house_hold_car_boat_mc_loan",
    "age",
    "requested_loan_over_gross_income",
    "register_ul_number_of_loans",
    "other_credits_limit",
    "n_applications_last_365_d",
]
data[numerical_cols] = imputer.fit_transform(data[numerical_cols])

# Normalize numerical data
scaler = MinMaxScaler()
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Handle categorical data
data = pd.get_dummies(data, columns=["employment_type"])

preprocessed_data = data
```

## 3. Propose Predictive Model with k-fold Cross-Validation:

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LogisticRegression


#features and target
X = data.drop(columns=['is_default_next_12_m'])
y = data['is_default_next_12_m']

#model with balanced class weights
model = LogisticRegression(class_weight='balanced')

kf = KFold(n_splits=5, shuffle=True, random_state=42)

#k-fold cross-validation
cv_scores = cross_val_score(model, X, y, cv=kf, scoring='roc_auc')
print("Cross-validation AUC-ROC scores:", cv_scores)
print("Mean AUC-ROC:", cv_scores.mean())
```

```python
from sklearn.model_selection import cross_val_predict, cross_val_score, KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, auc

X = data.drop(columns=['is_default_next_12_m'])
y = data['is_default_next_12_m']

#Random Forest model with balanced class weights
model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)

kf = KFold(n_splits=5, shuffle=True, random_state=42)

#generate predicted probabilities
predicted_probabilities = cross_val_predict(model, X, y, cv=kf, method='predict_proba')

y_scores = predicted_probabilities[:, 1]

fpr, tpr, thresholds = roc_curve(y, y_scores)
roc_auc = auc(fpr, tpr)

#ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve -  Random Forest')
plt.legend(loc='lower right')
plt.show()

model.fit(X, y)

feature_importances = model.feature_importances_
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
})

#df sort by importance values
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Feature Importance - Random Forest')
plt.show()
```
